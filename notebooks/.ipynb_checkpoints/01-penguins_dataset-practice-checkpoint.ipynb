```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.rcParams["figure.dpi"] = 120
```

```python
df = sns.load_dataset("penguins")
df
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe().T
```

```python
df["body_mass_kg"] = df["body_mass_g"] / 1000
df
```

```python
#df = df.fillna("")
df
```

```python
df = df.dropna(subset=["body_mass_g"]).reset_index(drop=True)
df
```

```python
df.groupby(["species", "island"], as_index=True).apply(lambda df: df)
```

```python
df.sort_values(by="body_mass_g", ascending=False)
```

```python
num_cols = df.select_dtypes(include=np.number).columns
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap="vlag", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.info()
```

```python
#category change
cat_cols = ["species", "island", "sex"]
for c in cat_cols:
    df[c] = df[c].astype("category")

df.info()
```

```python
df.select_dtypes(include=np.number).columns
```

```python
for n in num_cols:
    df[n] = df[n].astype("float64")

df.info()
```
